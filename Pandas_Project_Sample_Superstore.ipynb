Exploratory Data Analysis (EDA)

1-Top 10 most profit

```python
top_10_profit = ss.sort_values(by='Profit', ascending=False).head(10)
top_10_profit
```

2- show OrderID and Customer Name where is Sales > $1000

```python
ss[['Order ID', 'Customer Name','Sales']][ss['Sales'] > 1000].sort_values(by='Sales', ascending=False).head(10)
```

4- Which regions or customer segments should we focus on for growth?

South and Central by region

Home Office by Segments

```python
Total_Profit_Region = ss.groupby('Region')['Profit'].sum().sort_values(ascending=False)
Total_Profit_Segments= ss.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
print(Total_Profit_Region)
print(Total_Profit_Segments)
```

5- Total sales by region

```python
total_salesRegion = ss.groupby('Region')['Sales'].sum().sort_values(ascending=False)
total_salesRegion
```

6- Average profit by category

```python
avg_Profitbycategory= ss.groupby('Category')['Profit'].mean()
avg_Profitbycategory
```

7- Which product categories generate the most profit?

```python
category_profit = ss.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print("Product categories by total profit:")
category_profit
```

8- Which product category has the highest total sales?

```python
Highest_total_sales = ss.groupby('Category')['Sales'].sum().max()
Category_1 = ss.groupby('Category')['Sales'].sum().idxmax()
print(f'Product of category has the highest total sales is {Category_1} = {Highest_total_sales}')
```

9- Which 'Ship Mode' is used most frequently, and what is its average 'Ship Days'?

```python
Shop_mode = ss.groupby('Ship Mode') ['Ship Days'].mean()
Shop_mode
```

10- What is the total quantity of products sold for each 'Sub-Category'?

```python
Total_quantity = ss.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False)
Total_quantity
```

11- What is the category that is highest each year?

```python
#Highest_each_year = ss. groupby(['Year','Category']) ['Sales'].sum()
Highest_each_year = ss. groupby(['Year','Category']) ['Sales'].sum().groupby(level=0).idxmax()
Highest_each_year
```
